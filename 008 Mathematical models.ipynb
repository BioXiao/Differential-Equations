## Population growth

In some situation the rate of change in the number of individuals in a population at a specific time is directly proportional to the number of individuals at that specific time:
$$ \frac{dP}{dt} \propto P $$
where $ P $ is the population at time $ t $.

By introducting a constant of proportionality we have:
$$ \frac{dP}{dt} = kP $$
which we can solve.

$$ \frac { 1 }{ P } dP=kdt\\ \int { \frac { 1 }{ P }  } dP=\int { k } dt\\ \ln { \left| P \right|  } =k{ t }+c\\ \ln { \left| P \right|  } =k{ t }+c\\ P\left( t \right) ={ e }^{ k{ t }+c }={ e }^{ c }{ e }^{ k{ t } }\\ P\left( t \right) =c{ e }^{ k{ t } } $$

If $ P \left( 0 \right) = {P}_{0} $ we have:
$$ P\left( 0 \right) =c{ e }^{ \left( 0 \right)  }=c={ P }_{ 0 }\\ P\left( t \right) ={ P }_{ 0 }{ e }^{ k{ t } } $$

## Radioactive decay

Atomic nuclei consist of protons and neutrons.  Depending on the number of neutrons, some atomic nuclei are unstable and transmutations occurs deur to neutron decay.  Under some assumptions the rate of decay can be modeled as we had above with the rate of decay proprtional to the number of neuclei present.

This process has a half-life, which is similar to that of the elimination of some drugs from the human body.

The deriavtion of these equations of decay is exactly the same as [above](#Population-growth):
$$ P\left( t \right) ={ P }_{ 0 }{ e }^{ k{ t } } $$

It is only the value of $ k $ that chnages the equation from one of grwoth to one of decay, with $ k > 0 $ indicating growth and $ k < 0 $ indicating decay (negative growth).

## More generalized growth

The [population growth](#Population-growth) above did not take any *births* or *death* into consideration.  We can define birth and death rate funstions with $ \beta \left( t \right) $ the number of births per population unit per time unit and $ \delta \left( t \right) $ as the number of deaths per population unit per time unit.

Dring the time interval, $ \Delta t $, we will have the following number of births and deaths:
- Births = $ \beta \left( t \right) \cdot P \left( t \right) \cdot \Delta t $
- Deaths = $ \delta \left( t \right) \cdot P \left( t \right) \cdot \Delta t $

This make the change in the population number:
$$ \Delta P \approx \beta \left( t \right) \cdot P \left( t \right) \cdot \Delta t - \delta \left( t \right) \cdot P \left( t \right) \cdot \Delta t $$

The rate fo change in the population number is then:
$$ \frac { \Delta P }{ \Delta t } \approx \left( \beta \left( t \right) -\delta t \right) P\left( t \right) \\ \frac { dP }{ dt } =\left( \beta -\delta  \right) P $$

In a very simplistic scenario $ \beta $ and $ \delta $ are constant, i.e. $ k = \beta - \delta $ and we have the specific form we had above.

As long as $ \beta > \delta $, i.e. a $ k > 0 $ for constant values, we will always get that the population exands to infinity.  This is unrealistic.  So let's introduce a changing birth rate (but keep the death rate constant).

If we make the birth rate a linear function of the population number, i.e. $ {\beta}_{0} - {\beta}_{1} P $, with both values positive and we have $ \delta = {\delta}_{0} $ then we can derive what is called the **logistic equation**:
$$ \frac { dP }{ dt } =\left( { \beta  }_{ 0 }-{ \beta  }_{ 1 }P-{ \delta  }_{ 0 } \right) P\\ a={ \beta  }_{ 0 }-{ \delta  }_{ 0 },\quad b={ \beta  }_{ 1 }\\ \frac { dP }{ dt } =aP-b{ P }^{ 2 }\\ k=b,\quad M=\frac { a }{ b } \\ \frac { dP }{ dt } =kP\left( M-P \right)  $$

### Example 1

For $ k = 0.0004 $ and $ M = 150 $ calculate $ P \left( t \right) $:
$$ \frac { dP }{ dt } =0.0004P\left( 150-P \right)  $$

```python
x = linspace(0, 100, 2000)
pval = [20, 50, 100, 300]

plot(x, (150 * pval[0] * exp(0.06 * x)) / (150 - pval[0] + (pval[0] * exp(0.06 * x))), "b-")
plot(x, (150 * pval[1] * exp(0.06 * x)) / (150 - pval[1] + (pval[1] * exp(0.06 * x))), "b-")
plot(x, (150 * pval[2] * exp(0.06 * x)) / (150 - pval[2] + (pval[2] * exp(0.06 * x))), "b-")
plot(x, (150 * pval[3] * exp(0.06 * x)) / (150 - pval[3] + (pval[3] * exp(0.06 * x))), "b-")
show();
```
